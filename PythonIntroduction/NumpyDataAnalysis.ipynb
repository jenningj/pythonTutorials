```python
#arr.astype(np.float64)
#sometimes it will not owrk
points = np.arange(-5,5,0.01)
```

```python
xs,ys = np.meshgrid(points,points)
```

```python
ys
```

```python
import matplotlib.pyplot as plt
z = np.sqrt(xs **2 + ys ** 2)
```

```python
z
```

```python
%matplotlib inline
plt.title("Image plot of $\sqrt{x^r + y^2}$ for a frid of values")
plt.imshow(z,cmap = plt.cm.gray);plt.colorbar()
```

```python
#Use np.where instead of list comprehension
np.where(z > 0, 5, z) # set only positive values to 2
plt.imshow(z,cmap = plt.cm.gray);plt.colorbar()
```
